A rather off-topic yet interesting usage for clustering is compressing images! Standard raw images come in RGB format with 32 bits (3 bytes) for each pixel. A 1280 x 720 raw image requires 1280*720*3 = 2764800 bytes = ~2.8 MB. However, most common image formats (JPEG, PNG, etc.) take much less space because they are in a (lossy or lossless) compressed form.
One lossy algorithm can be implemented with k-means! The idea is to decrease the number of colours in the image which enables decreasing the number of bits needed for each pixel. Clustering the colours together can give a representative of each group of similar colors (the cluster center) to assign them to. This type of compression is called Vector Quantization.

# Download and read the cat image 

```python
import numpy as np
import pandas as pd
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
im = mpimg.imread('cat.png')
imgplot = plt.imshow(im)
plt.show()
```

```python
type(im)
```

```python
im_height = im.shape[0]
im_width = im.shape[1]
```

```python
im.shape
```

# Re-organize
Re-organize the pixels as a 2D array with dimensions (number of pixels, 3).
In other words, a dataset with number of pixels samples and 3 columns.

```python
im_arr = np.reshape(im, (im.shape[0]*im.shape[1], 3))
```

```python
im_arr.shape
```

```python
# display as df
im_df=pd.DataFrame(im_arr, columns=["R", "G", "B"])
```

```python
im_df.describe()
```

```python
im_df.head()
```

# Clustering
Use K-means to cluster the data into 10 clusters.
Replace each pixel by its cluster center.
Re-organize the pixels into a 3D array to show the image.


```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10, tol = 0.01)
clusters_fit = kmeans.fit(im_df)
clusters = kmeans.predict(im_df)
```

```python
# create an array replacing each pixel label with its corresponding cluster centroid
clusters_arr = np.array([list(clusters_fit.cluster_centers_[label]) for label in clusters])
clusters_arr
```

```python
clusters_arr = (clusters_arr*256).astype("uint8")
compressed_img = np.reshape(clusters_arr, (im_height, im_width, 3), "C")
```

```python
plt.imshow(compressed_img)
plt.show()
```

# Question:
What happens to the compressed image if k increases?

```python
num_clusters=[2, 5, 10, 50, 100]

fig, axs = plt.subplots(1,5, figsize=(15, 10), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .5, wspace=.001)

axs = axs.ravel()
loc=0
for i in num_clusters:
  
    kmeans = KMeans(n_clusters=i, tol = 0.01)
    clusters_fit = kmeans.fit(im_df)
    clusters = kmeans.predict(im_df)
    #clusters_i.append(clusters)
    
    #clusters_arr_i.append(np.array([list(clusters_fit.cluster_centers_[label]) for label in clusters]))
    clusters_arr = np.array([list(clusters_fit.cluster_centers_[label]) for label in clusters])
    clusters_int = (clusters_arr *256).astype("uint8")
    compressed_img_i = np.reshape(clusters_int, (im_height, im_width, 3), "C")
    
    axs[loc].imshow(compressed_img_i)
    axs[loc].set_title(str(num_clusters[loc]))
    name="cat"+str(i)+".png"
    #plt.imsave(name, compressed_img_i) 
    
    loc=loc+1
```

```python
# As k increased the image quality increases and allows for a larger range of colours. 
# This also creates a larger file size 
```
